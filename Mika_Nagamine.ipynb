```python
 # Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read Population-EstimatesData into pandas
pop_df=pd.read_csv('/Users/mikan/Documents/Data Science Bootcamp/populations_project/Population-EstimatesData.csv')
pop_df.head()
```

```python
# Select the necessary rows and columns
# Rows - European Union, Central Europe and the Baltics, Sub Saharan Africa, United States, Latin America and Caribbean, 
# Middle East and North Africa, East Asia and Pacific, South Asia, World
# Columns Country Name, Country Code, Indicator Name, Indicator Code, 1970 - 2050

pop_df=pop_df.rename(columns={'Country Name':'Country_Name',
                     'Country Code':'Country_Code',
                     'Indicator Name':'Indicator_Name',
                     'Indicator Code':'Indicator_Code'})
pop_df.head()
```

```python
# Select columns that are necessary 
data1970_df=pop_df.loc[:, ['Country_Name','Indicator_Name', '1970']]
data1970_df
```

```python
# Select countries / regions
names = ['European Union', 'Central Europe and the Baltics', 'Sub-Saharan Africa', 'United States',
        'Latin America & Caribbean', 'Middle East & North Africa', 'East Asia & Pacific', 'South Asia']

world = ['World']

data1970_df2= data1970_df[data1970_df.Country_Name.isin(names)]
data1970_df2.head()
```

```python
# Save the population world wide in 1970, 2020, 2050 for the later use
# 1970
world_70=data1970_df[data1970_df.Country_Name.isin(world)]
world_70

# Select population total row from World
world_70_1=world_70.loc[world_70['Indicator_Name'] == 'Population, total']
world_70_1
```

```python
# Select population total row from each region
data1970_df3=data1970_df2.loc[data1970_df['Indicator_Name'] == 'Population, total']
data1970_df3
```

```python
# Create the similar table for 2020

# Select columns that are necessary 
data2020_df=pop_df.loc[:, ['Country_Name','Indicator_Name', '2020']]
data2020_df

# Select necessary regions / country names
names = ['European Union', 'Central Europe and the Baltics', 'Sub-Saharan Africa', 'United States',
        'Latin America & Caribbean', 'Middle East & North Africa', 'East Asia & Pacific', 'South Asia']

data2020_df2= data2020_df[data2020_df.Country_Name.isin(names)]
data2020_df2.head()

# Select necessary indicator name
data2020_df3=data2020_df2.loc[data2020_df['Indicator_Name'] == 'Population, total']
data2020_df3
```

```python
# save the worldpopulation in 2020 for later use
world_20=data2020_df[data2020_df.Country_Name.isin(world)]
world_20

# Select population total row from World
world_20_1=world_20.loc[world_20['Indicator_Name'] == 'Population, total']
world_20_1
```

```python
# Create the similar table for 2050

# Select columns that are necessary 
data2050_df=pop_df.loc[:, ['Country_Name','Indicator_Name', '2050']]
data2050_df

# Select necessary regions / country names
names = ['European Union', 'Central Europe and the Baltics', 'Sub-Saharan Africa', 'United States',
        'Latin America & Caribbean', 'Middle East & North Africa', 'East Asia & Pacific', 'South Asia']

data2050_df2= data2050_df[data2050_df.Country_Name.isin(names)]
data2050_df2.head()

# Select necessary indicator name
data2050_df3=data2050_df2.loc[data2050_df['Indicator_Name'] == 'Population, total']
data2050_df3
```

```python
# save the worldpopulation in 2050 for later use
world_50=data2050_df[data2050_df.Country_Name.isin(world)]
world_50

# Select population total row from World
world_50_1=world_50.loc[world_50['Indicator_Name'] == 'Population, total']
world_50_1
```

```python
# Calculate total population in the selected regions in 1970, 2020, 2050

total_pop70 = int(data1970_df3["1970"].sum())
total_pop70 # 3465294538

total_pop20 = int(data2020_df3["2020"].sum())
total_pop20 # 7409282000

total_pop50 = int(data2050_df3["2050"].sum())
total_pop50  # 9293664000

# Calculate sum of population for each region for each year
percentage_pop70 = data1970_df3.groupby(["Country_Name", "Indicator_Name"]).sum() 
percentage_pop70
```

--- Summary --- 

The goal of this section is to present the overall changes in population distribution in the selected 8 regions from the past to the future. 

The selected regions are below: 
Central Europe and the Baltics
East Asia & Pacific
European Union
Latin America & Caribbean
Middle East & North Africa
South Asia
Sub-Saharan Africa
United States

Overall, about 94% - 96% of the world population is in the 8 regions any time between 1970 and 2050. For example, about 7.8 billion people are on Earth in 2020. Among them, 7.4 billion are in these 8 regions.  

Population in 1970 was concentrated in either Asian countries (58%) and European countries (13%). European Union was the 3rd most populated area on Earth in 1970 followed by Latin America and Caribbean. In 2020, population seems to have declined within the majority of the regions except in Sub-Saharan Africa and South Asia. The proportion of the population in European Union is currently the 5th out of 8 regions. This could indicate the population European countries declined in the last 50 years. In the future (2050), Sub-Saharan African countries are expected to have about 1/4 of population. Asian countries seem to have about 1/2 of the remaining population. In other words, 3/4 of the population is expected either in Asia or Sub-Saharan African countries. Total population of the rest of 5 regions is expected to have about 25% of the whole population. European Union is one of these 5 regions, the 3rd from the least populated region  and will have only about 5.4% of all people on Earth. 

1970:
In 1970, the population was concentrated in mainly Asian countries followed by European Union: 58% in East Asia & Pacific and South Asian countries (37%, 21% respectively); 13% in European Union. 

```python
# Generate pie chart - 1970
pies = ["Central Europe and the Baltics", "East Asia & Pacific", "European Union", "Latin America & Caribbean", 
        "Middle East & North Africa", "South Asia", "Sub-Saharan Africa", "United States"]
pie_pop = [9.913455e+07,1.290384e+09,4.419862e+08,2.860272e+08,
           1.384731e+08,7.137113e+08,2.905262e+08,2.050520e+08]
explode = (0,0.1,0,0,0,0,0,0)
colors = ["blue", "green", "red", "cyan", "magenta", "yellow", "orange", "purple"]

fig1, ax1 = plt.subplots()
ax1.pie(pie_pop, explode=explode, autopct='%1.1f%%', textprops=dict(color="black"), colors=colors,
        shadow=True, startangle=90)

ax1.legend(pies,
          title="Regions",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.3, 1))

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


ax1.set_title(f"1970 Population Distribution\n Total Population in 8 Regions: {total_pop70:,.0f}")

plt.savefig("/Users/mikan/Documents/Data Science Bootcamp/populations_project/Pie1970.png",bbox_inches='tight')

plt.show()
```

2020:

In 2020, Asian countries are still a population center. About 57% of the population are living in East Asia & Pacific and South Asian countries. This trend did not seem to have changed since 1970, but the distributions of people within these countries have changed.  East Asia & Pacific countries still have 31% of population, but this proportion is 6% smaller than 1970. On the other hand, South Asian countries have 25% of population that is 4% larger than 1970.

Population in European Union seems to have declined. It is only 7% of all the population in the 8 regions and no longer the 3rd most populated area. 

A dramatic increase in population seems to have occurred in Sub-Saharan African countries in the last 50 years. This region became to have 15% of all the population and the third most populated areas on Earth in 2020.
